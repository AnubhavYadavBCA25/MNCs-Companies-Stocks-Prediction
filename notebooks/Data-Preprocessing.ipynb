### 4. Feature Engineering

**Logarithmic Returns:** Logarithmic returns, often simply called "log returns," are a way of calculating returns on an investment, particularly useful in financial modeling and time series analysis.  They are calculated as the natural logarithm of the ratio of the ending price to the beginning price:  

```Log Return = ln(Ending Price / Beginning Price)```

Where ```ln``` is the natural logarithm.

````python
# Logarithmic Returns for each stock
df_amazon['Log_Returns'] = np.log(df_amazon['Close']/df_amazon['Close'].shift(1))
df_apple['Log_Returns'] = np.log(df_apple['Close']/df_apple['Close'].shift(1))
df_google['Log_Returns'] = np.log(df_google['Close']/df_google['Close'].shift(1))
df_meta['Log_Returns'] = np.log(df_meta['Close']/df_meta['Close'].shift(1))
df_micro['Log_Returns'] = np.log(df_micro['Close']/df_micro['Close'].shift(1))
df_netflix['Log_Returns'] = np.log(df_netflix['Close']/df_netflix['Close'].shift(1))
df_nvidia['Log_Returns'] = np.log(df_nvidia['Close']/df_nvidia['Close'].shift(1))
````

#### EMA 10: 
**Exponential Moving Average (EMA):** An EMA is a type of moving average that gives more weight and significance to recent price data. This makes it more responsive to new price changes compared to a Simple Moving Average (SMA), which treats all data points equally.

**10-day:** This specifies the period over which the EMA is calculated. In this case, it means the EMA is based on the closing prices of the last 10 trading days.

#### How to calculate:
1. **Calculate the Simple Moving Average (SMA):** First, calculate the SMA for the initial 10 days. This is simply the sum of the closing prices for those 10 days divided by 10.

2. **Calculate the Multiplier:** The multiplier determines how much weight is given to the most recent price. It's calculated as: ```Multiplier = 2 / (Period + 1)``` In this case, the multiplier for a 10-day EMA is 2 / (10 + 1) = 0.1818 (approximately).

3. **Calculate the EMA:** The EMA is calculated using the following formula:

    ```EMA = (Closing Price * Multiplier) + (Previous EMA * (1 - Multiplier))```

````python
# EMAs for each stock
df_amazon['EMA_10'] = df_amazon['Close'].ewm(span=10, adjust=False).mean()
df_apple['EMA_10'] = df_apple['Close'].ewm(span=10, adjust=False).mean()
df_google['EMA_10'] = df_google['Close'].ewm(span=10, adjust=False).mean()
df_meta['EMA_10'] = df_meta['Close'].ewm(span=10, adjust=False).mean()
df_micro['EMA_10'] = df_micro['Close'].ewm(span=10, adjust=False).mean()
df_netflix['EMA_10'] = df_netflix['Close'].ewm(span=10, adjust=False).mean()
df_nvidia['EMA_10'] = df_nvidia['Close'].ewm(span=10, adjust=False).mean()
````

#### RSI
**RSI:** RSI stands for the Relative Strength Index. It's a popular momentum indicator used in technical analysis to measure the speed and change of price movements. Think of it like a speedometer for stock prices, showing how quickly and strongly they are changing.

#### What it is:
1. **Momentum Oscillator:** The RSI is an oscillator, meaning it fluctuates between two extremes. In this case, the values range from 0 to 100.
2. **Measures Momentum:** It gauges the momentum of price changes, helping to identify overbought and oversold conditions.
3. **14-day Period:** The RSI is typically calculated over a 14-day period, though this can be adjusted.

#### How it works:
1. **Calculate Average Gains and Losses:** The RSI calculation involves comparing the average gains and losses over the chosen period (e.g., 14 days).
2. **Formula:** The formula to calculate RSI is a bit complex, but here's the gist:

    ```RSI = 100 - (100 / (1 + RS))```

    Where RS is the average of upward price changes divided by the average of downward price changes.

3. **Values:** The resulting RSI value falls between 0 and 100.

````python
# RSI for each stock
def RSI(data, window=14):
    delta = data['Close'].diff(1)
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    RS = gain / loss
    return 100 - (100 / (1 + RS))

df_amazon['RSI'] = RSI(df_amazon)
df_apple['RSI'] = RSI(df_apple)
df_google['RSI'] = RSI(df_google)
df_meta['RSI'] = RSI(df_meta)
df_micro['RSI'] = RSI(df_micro)
df_netflix['RSI'] = RSI(df_netflix)
df_nvidia['RSI'] = RSI(df_nvidia)
````

````python
# Replace RSIs with bfilled values
df_amazon['RSI'] = df_amazon['RSI'].fillna(method='bfill')
df_apple['RSI'] = df_apple['RSI'].fillna(method='bfill')
df_google['RSI'] = df_google['RSI'].fillna(method='bfill')
df_meta['RSI'] = df_meta['RSI'].fillna(method='bfill')
df_micro['RSI'] = df_micro['RSI'].fillna(method='bfill')
df_netflix['RSI'] = df_netflix['RSI'].fillna(method='bfill')
df_nvidia['RSI'] = df_nvidia['RSI'].fillna(method='bfill')
````

#### EMA 12 and EMA 26

- **EMA 12:** This is a relatively short-term EMA, meaning it reacts quickly to recent price changes. It's calculated using the closing prices of the last 12 trading days, giving more weight to the most recent prices.  

- **EMA 26:** This is a medium-term EMA, reacting a bit slower to price changes compared to the EMA 12. It's calculated using the closing prices of the last 26 trading days, again with more weight on recent prices.

````python
# MACD for each stock

# Calculate the short term and long term exponential moving average (EMA_12, EMA_26)
df_amazon['EMA_12'] = df_amazon['Close'].ewm(span=12, adjust=False).mean()
df_amazon['EMA_26'] = df_amazon['Close'].ewm(span=26, adjust=False).mean()

df_apple['EMA_12'] = df_apple['Close'].ewm(span=12, adjust=False).mean()
df_apple['EMA_26'] = df_apple['Close'].ewm(span=26, adjust=False).mean()

df_google['EMA_12'] = df_google['Close'].ewm(span=12, adjust=False).mean()
df_google['EMA_26'] = df_google['Close'].ewm(span=26, adjust=False).mean()

df_meta['EMA_12'] = df_meta['Close'].ewm(span=12, adjust=False).mean()
df_meta['EMA_26'] = df_meta['Close'].ewm(span=26, adjust=False).mean()

df_micro['EMA_12'] = df_micro['Close'].ewm(span=12, adjust=False).mean()
df_micro['EMA_26'] = df_micro['Close'].ewm(span=26, adjust=False).mean()

df_netflix['EMA_12'] = df_netflix['Close'].ewm(span=12, adjust=False).mean()
df_netflix['EMA_26'] = df_netflix['Close'].ewm(span=26, adjust=False).mean()

df_nvidia['EMA_12'] = df_nvidia['Close'].ewm(span=12, adjust=False).mean()
df_nvidia['EMA_26'] = df_nvidia['Close'].ewm(span=26, adjust=False).mean()
````

#### MACD
The Moving Average Convergence Divergence (MACD) is a momentum indicator that shows the relationship between two moving averages of a security's price. It's a popular tool used by traders to identify trends, potential buy and sell signals, and overbought/oversold conditions.

#### How to calculate
```MACD Line = 12-day EMA - 26-day EMA```

````python
# Calculate the MACD line
df_amazon['MACD'] = df_amazon['EMA_12'] - df_amazon['EMA_26']
df_apple['MACD'] = df_apple['EMA_12'] - df_apple['EMA_26']
df_google['MACD'] = df_google['EMA_12'] - df_google['EMA_26']
df_meta['MACD'] = df_meta['EMA_12'] - df_meta['EMA_26']
df_micro['MACD'] = df_micro['EMA_12'] - df_micro['EMA_26']
df_netflix['MACD'] = df_netflix['EMA_12'] - df_netflix['EMA_26']
df_nvidia['MACD'] = df_nvidia['EMA_12'] - df_nvidia['EMA_26']
````

#### Signal Line
The signal line is a 9-day Exponential Moving Average (EMA) of the MACD line itself. Think of it as a smoothed-out version of the MACD line. It helps to identify potential turning points and generate trading signals.

#### How to Calculate the Signal Line

1. **Gather MACD Line Values:** You'll need the calculated MACD line values for the last 9 periods (usually days).

2. **Calculate the Multiplier:** The multiplier for the signal line (which is a 9-day EMA) is:

    ```Multiplier = 2 / (Period + 1) = 2 / (9 + 1) = 0.2```

3. **Calculate the Signal Line:** The formula for calculating the signal line is the same EMA formula we used for the 12-day and 26-day EMAs, but now we're applying it to the MACD line values:

    ```Signal Line = (Current MACD Line Value * Multiplier) + (Previous Signal Line Value * (1 - Multiplier))```

````python
# Calculate the Signal Line
df_amazon['Signal_Line'] = df_amazon['MACD'].ewm(span=9, adjust=False).mean()
df_apple['Signal_Line'] = df_apple['MACD'].ewm(span=9, adjust=False).mean()
df_google['Signal_Line'] = df_google['MACD'].ewm(span=9, adjust=False).mean()
df_meta['Signal_Line'] = df_meta['MACD'].ewm(span=9, adjust=False).mean()
df_micro['Signal_Line'] = df_micro['MACD'].ewm(span=9, adjust=False).mean()
df_netflix['Signal_Line'] = df_netflix['MACD'].ewm(span=9, adjust=False).mean()
df_nvidia['Signal_Line'] = df_nvidia['MACD'].ewm(span=9, adjust=False).mean()
````
